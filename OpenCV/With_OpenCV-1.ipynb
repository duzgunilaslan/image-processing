## Resizing & Rescaling

```python
def rescaleFrame(frame, scale = 0.50):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimantions = (width,height)
    
    return cv.resize(frame, dimantions, interpolation = cv.INTER_AREA)

rimg =rescaleFrame(img)
cv.imshow("resize", rimg)
cv.waitKey(0)
```

```python
#Cropping 
cropped = img[50:200, 200:400]
cv.imshow("Cropped",cropped)
cv.waitKey(0)
```

## Image Transformations

```python
def translate(img, x, y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimantions = (img.shape[1], img.shape[0])
    return cv.warpAffine(img, transMat, dimantions)

# -x --> shift left
# -y --> shift right
#  x --> shift right
#  y --> shift left

translated = translate(img, 150, 150)
cv.imshow("translate", translated)
cv.waitKey(0)
```

```python
# Rotation
def rotate(img, angle, rotPoint = None):
    (height,width) = img.shape[:2]
    
    if rotPoint is None:
        rotPoint = (width//2, height//2)
        
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0)
    dimensions = (width, height)
    
    return cv.warpAffine(img, rotMat, dimensions)

rotated = rotate(img, 45)
cv.imshow("Rotated", rotated)
cv.waitKey(0)
```

## Contour Detections

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("Gray", gray)
cv.waitKey(0)
```

```python
blank = np.zeros(img.shape, dtype = "uint8")
ret, thresh = cv.threshold(gray, 125, 125, cv.THRESH_BINARY)
contours, hierarchies = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_SIMPLE)
print(f"{len(contours)} contour(s) found !")
cv.drawContours(blank, contours, -1, (0,0,255), 1)
cv.imshow("Contours Draw",blank)
cv.waitKey(0)
```
