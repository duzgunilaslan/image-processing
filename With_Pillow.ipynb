## Image Enhancement

Sharpness Enhancement
Contrast Enhancement
Brightness Enhancement
Color Enhancement

```python
from PIL import ImageEnhance

plt.figure(figsize = (10,10))
image_color_enhan = image.copy()
image1 = ImageEnhance.Color(image_color_enhan).enhance(2.5)
image2 = ImageEnhance.Contrast(image_color_enhan).enhance(2.5)
image3 = ImageEnhance.Brightness(image_color_enhan).enhance(1.5)
image4 = ImageEnhance.Sharpness(image_color_enhan).enhance(2.5)

plt.subplot(2,2,1)
plt.imshow(image1)
plt.title("Color")


plt.subplot(2,2,2)
plt.imshow(image2)
plt.title("Contrast")

plt.subplot(2,2,3)
plt.imshow(image3)
plt.title("Brightness")

plt.subplot(2,2,4)
plt.imshow(image4)
plt.title("Sharpness")
```

```python
# Reduce the enhance values


plt.figure(figsize = (10,10))
image_color_enhan = image.copy()
image1 = ImageEnhance.Color(image_color_enhan).enhance(.5)
image2 = ImageEnhance.Contrast(image_color_enhan).enhance(.5)
image3 = ImageEnhance.Brightness(image_color_enhan).enhance(.5)
image4 = ImageEnhance.Sharpness(image_color_enhan).enhance(.5)

plt.subplot(2,2,1)
plt.imshow(image1)
plt.title("Color")


plt.subplot(2,2,2)
plt.imshow(image2)
plt.title("Contrast")

plt.subplot(2,2,3)
plt.imshow(image3)
plt.title("Brightness")

plt.subplot(2,2,4)
plt.imshow(image4)
plt.title("Sharpness")
```

## Alpha Blending

Alpha blending is the mixing of two images. It is combinig of the alpha channel 
with other layers in an image in order to show translucency.

out = image1 * (1.0 - alpha) + image2 * alpha


Notice:
    - Make sure the images have the same size
    - Make sure the images have the alpha channel (*png format) 

```python
image1 = image.copy()
image2 = image2.copy()

#making both images same size
image2 = image2.resize(image1.size)

#Blending

image_blend = Image.blend(image1, image2, .5)
plt.imshow(image_blend)
```
